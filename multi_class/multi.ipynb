```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns 
import joblib 
import sklearn 
```

```python
df1=pd.read_csv('Iris.csv')
df=df1.copy()
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
for i in df.select_dtypes(include='number').columns:
  sns.boxplot(df[i])
  plt.show()
```

```python
def out(col):
  q1,q3=np.percentile(col,[25,75])
  iqr=q3-q1
  ll=q1-1.5*iqr 
  ul=q3+1.5*iqr 
  return ll,ul 
for i in df.select_dtypes(include='number').columns:
  ll,ul=out(df[i])
  df[i]=np.where(df[i]<ll,ll,df[i])
  df[i]=np.where(df[i]>ul,ul,df[i])
```

```python
for i in df.select_dtypes(include='number').columns:
  sns.boxplot(df[i])
  plt.show()
```

```python
df.head()
```

```python
df['Species'] = df['Species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2})

# Fill any unmapped values with a default (e.g., -1) before converting to int
df['Species'] = df['Species'].fillna(-1).astype(int)
```
